```python
#help function
from datetime import datetime, timedelta

def add_days_to_date(date_str, days):
    # Parse the date string into a datetime object
    date_obj = datetime.strptime(date_str, "%Y-%m-%d")
    
    # Add the desired number of days
    new_date_obj = date_obj + timedelta(days=days)
    
    # Convert the new datetime object back to a string
    new_date_str = new_date_obj.strftime("%Y-%m-%d")
    
    return new_date_str
```

```python
def get_moving_average(stock_symbol,rise_date, window=20):
    # Fetch historical data
    stock_data = yf.download(stock_symbol, start=add_days_to_date(rise_date, -36), end=rise_date, interval="1d")
    
    # Calculate the moving average
    stock_data['MA20'] = stock_data['Close'].rolling(window=window).mean()
    
    return stock_data
```

```python
def is_ma20_going_up(stock_symbol, rise_date): #when the stock is moving up at this date
    stock_data = get_moving_average(stock_symbol, rise_date)
    
    # Check if the last two MA20 values are increasing
    if stock_data['MA20'].iloc[-1] > stock_data['MA20'].iloc[-2]:
        if stock_data['Close'].iloc[-1] > stock_data['MA20'].iloc[-1]:
            return True
        else:
            return False
    else:
        return False
```
